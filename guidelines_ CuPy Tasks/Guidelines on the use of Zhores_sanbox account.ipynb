## How to <i><u>get into</u></i> Zhores_sandbox account?

### Credentials
* Login,
* Password. 

#### !Please donâ€™t forget to change your password using 'passwd' command after the first login!

* Login node name: <strong>sandbox.zhores.net</strong>
* SSH port: <strong>2200</strong>
* ```bash
[local@computer ...]$ ssh -p 2200 n.surname@sandbox.zhores.net
```

### Workload manager
* The <strong> Slurm </strong> workload manager is installed on the cluster.
* To allocate <strong>cpu </strong> node:
```bash
[n.surname@an ~]$ salloc -p cpu
```

* To allocate $<$ number_of_gpus $>$ <strong>gpus </strong>:
```bash
[n.surname@an ~]$ salloc -p gpu --nodes=<number_of_nodes> --gpus=<number_of_gpus> --gpus-per-node=<number_of_gpus_per_node>
```

* To get information on nodes:
```bash
[n.surname@an ~]$ sinfo
```

* To get information on queues:
```bash
[n.surname@an ~]$ squeue
```

* To get information on your jobs:
```bash
[n.surname@an ~]$ squeue -u n.surname
```

* To cancel the job:
```bash
[n.surname@an ~]$ scancel <job_number>
```


### Storage

### Software modules
* To list all available module do:
```bash
[n.surname@an ~]$ module avail
```

* To load module do:
```bash
[n.surname@an ~]$ module load <module_name>
```

* To list loaded modules:
```bash
[n.surname@an ~]$ module list
```

## How to install <u><i>cupy</i></u>?

1. 
```bash
[n.surname@an ~]$ module load gpu/cuda-10.0
```

2. 
```bash
[n.surname@an ~]$ module load python/python-3.7.1
```

3. 
```bash
[n.surname@an ~]$ pip3.7 install cupy-cuda100 --user
```

## How to launch the <u><i>cupy</i></u>-based python file <u><i>file.py</i></u>?

1. 
```bash
[n.surname@an ~]$ salloc -p gpu --gpus=1
```

2. <u>If you didn't load cuda-10.0:</u>
```bash
[n.surname@an ~]$ module load gpu/cuda-10.0
```

3. <u>If you didn't load python-3.7.1:</u>
```bash
[n.surname@an ~]$ module load python/python-3.7.1
```

4. 
```bash
[n.surname@an ~]$ srun python3.7 file.py
```

5. 
```bash
[n.surname@an ~]$ scancel <job_number>
```

## How to conduct simultaneous monitoring of resources?
<big> Resources monitoring of memory (such as GPU m. and RAM), GPUs and CPUs load, volatile GPU-utilization, power usage. </big>

1. Open additional ssh connection to Zhores_sandbox by using extra terminal.
2. 

    1. If you allocated gpus: 
```bash
[n.surname@an ~]$ ssh gn<gpu_node_number>
```

        * Monitor gpu resources:
```bash
[n.surname@gn<gpu_node_number> ~]$ watch -n 0,1 nvidia-smi
```

    2. If you allocated cpu:
```bash
[n.surname@cn<cpu_node_number> ~]$ ssh cn<cpu_node_number>
```

3. 
```bash
[n.surname@...n... ~]$ top
```

4. 
```bash
[n.surname@...n... ~]$ htop
```
